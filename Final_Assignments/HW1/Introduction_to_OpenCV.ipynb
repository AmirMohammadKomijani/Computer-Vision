Mode 1 : The main image should be joined vertically with its vertical invert.

```python
img = cv2.imread('ComputerDepartment.jpg') ## i read image with new variable
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB) ## convert img to RGB manner because i undrestand RGB easier.
```

```python
## source : https://www.tutorialspoint.com/how-to-flip-an-image-in-opencv-python#:~:text=In%20OpenCV%2C%20an%20image%20can,the%20image%20across%20the%20axis.
## source : https://stackoverflow.com/questions/7589012/combining-two-images-with-opencv
flipped_img = cv2.flip(img,0)  ## flipping image vertically
result = np.concatenate((img, flipped_img), axis=0) ## concatenate these two images
plt.imshow(result)
```

Mode 2 : The vertical invert of the original image should be joined vertically with the original image.

```python
result = np.concatenate((flipped_img, img), axis=0) ## it is same as MODE1 , i just change the position of flipped_img and img
plt.imshow(result)
```

Mode 3 : The main image should be joined horizontally with its horizontal invert.

```python
flipped_img = cv2.flip(img,+1)
result = np.concatenate((img, flipped_img), axis=1) ## axis 1 for horizontally
plt.imshow(result)
```

Mode 4 : The horizontal invert of the original image should be joined horizontally with the original image.

```python
flipped_img = cv2.flip(img,+1)
result = np.concatenate((flipped_img, img), axis=1) ## change the flipped_img and img position in concatenate function
plt.imshow(result)
```

First, read and display the image of the iust logo. Then show each of its color channels so that the ​blue channel​ is displayed in the blue, ​green channel​ in the green, and ​red channel​ in red. Notice that since these are color images they have to be ​m​x​n​x​3 ​arrays. For a purely red image, the green and blue channels are entirely zero.

Read logo and show it.

```python
logo = cv2.imread('logo.jpg')
logo = cv2.cvtColor(logo,cv2.COLOR_BGR2RGB) ## convert to rgb mode for undrestanding easily
plt.imshow(logo)
```

Show blue channel.

```python
## source : https://stackoverflow.com/questions/44554125/python-want-to-display-red-channel-only-in-opencv
blue = logo.copy()
blue[:,:,0]=0 ## disable red channel with give 0 value to it
blue[:,:,1]=0 ## disable green channel with give 0 value to it
plt.imshow(blue)
```

Show green channel.

```python
green = logo.copy()
green[:,:,0]=0
green[:,:,2]=0
plt.imshow(green)
```

Show red channel.

```python
red = logo.copy()
red[:,:,1]=0
red[:,:,2]=0
plt.imshow(red)
```
