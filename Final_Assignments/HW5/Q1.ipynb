```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Read the image
image = cv2.imread('images/q1.png')
if image is None:
    raise ValueError("Image not found or the path is incorrect")

# Step 2: Convert to gray
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)


# Step 3: Find connected components
num_labels, labels_im= cv2.connectedComponents(gray)

# Step 4: Create a random color map
colors = [np.random.randint(0, 255, size=3).tolist() for _ in range(num_labels)]

# Step 5: Apply colors to the connected components
output_image = np.zeros_like(image)
for label in range(1, num_labels):
    output_image[labels_im == label] = colors[label]

# Step 6: Print the number of components on the image
text = f'Total components: {num_labels - 1}'
cv2.putText(output_image, text, (10, 30), cv2.FONT_HERSHEY_SIMPLEX, 1, (255, 255, 255), 2)

# Display the image using matplotlib
plt.figure(figsize=(10, 10))
plt.imshow(cv2.cvtColor(output_image, cv2.COLOR_BGR2RGB))
plt.title('Connected Components')
plt.axis('off')
plt.show()

# Save the output image
cv2.imwrite('output_image.png', output_image)
```
